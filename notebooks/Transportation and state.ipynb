```python
# use the import keyword to import pandas, requests, and bs4 modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# imported matplot for data ploting options for graph 
```

```python
# import data from file 
data=pd.read_csv("../data/2020_TransBorder_Data_State_and_Commodity.csv")
```

# Transportation state and commodity analysis 

```python
# checking size of data set in rows and columns 
data.head()
```

```python
data
```

```python
# create data frame for average freight charges by state for year 2020

sfreights = pd.DataFrame(data.groupby('USASTATE')['FREIGHT_CHARGES'].mean().sort_values(ascending=False))
sfreights
```

```python
# round the numbers up 
sfreights.round(1).reset_index()
```

```python
# output dataframe to a csv
sfreights.to_csv('../data/state_averages.csv', index=True)
```

```python
# create new data frame to show freight charge changes from January to fnruary by state 
na = pd.DataFrame(data.groupby(by=['USASTATE','MONTH'])['FREIGHT_CHARGES'].mean()).reset_index()
na
```

```python
na= na.pivot(index='USASTATE',columns='MONTH', values='FREIGHT_CHARGES')
```

```python
na
```

```python
na.to_csv('../data/state_averages_month.csv', index=True)
```
